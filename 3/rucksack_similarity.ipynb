```python
def get_group_score(group_list):

    common= group_list[0]
    for i in range(len(group_list)-1):
        
        next_wrd=group_list[i+1]

        common=list(set(common).intersection(set(next_wrd)))
    
    
    return char_dict[common[0]]
        

```

```python
def get_scores(file='sample.txt'):
    
    
    total_score=0
    
    with open(file) as f:
        lines = f.readlines()

    
        line_counter=0

        grp_list=[]
        for line in lines:
            
            if line == '\n':
                if len(grp_list)==3:
                    
                    total_score=total_score+get_group_score(grp_list)
                grp_list=[]
                line_counter=0
                print('end',grp_list,total_score)
                
            else:
                line_counter+=1
                print('line_counter:',line_counter)
                
                if(line_counter==4):

                    print(grp_list,total_score)
               
                    total_score=total_score+get_group_score(grp_list)
                    grp_list=[]
                    line_counter=1
                    grp_list.append(line.strip())
                    print(grp_list,total_score)
                else:
                    grp_list.append(line.strip())

                print(total_score)

    return total_score

                
                #                                     choices[1]
                                                
                

            
    
    # return total_score
```
