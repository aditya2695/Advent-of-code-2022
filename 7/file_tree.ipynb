```python
from bigtree import BaseNode,Node, print_tree, find,find_name,find_names, findall,find_attr
```

```python
class FileNode(Node):

    # def __init__(self, name: str, size: int = 0, **kwargs):
    #     super().__init__(name, **kwargs)
    #     self._size = size
        

    def __init__(self,name:str,type='',size=0,**kwargs):
        super().__init__(name, **kwargs)
        self._type=type
        self._size=size   
        #self.id=self.name+next(self.ancestors).name

    @property
    def type(self):
        return self._type
    
    @property
    def size(self):
        if self.is_leaf:
            return self._size
        total=sum([child.size for child in self.children])

        return total #if total<=100000 and self._type == 'dir' else 0



    
```

```python
def read_Treedata(file='sample.txt'):
    
    
    root=FileNode(name='root',type='dir')
    current=root
    
    listing=False
    with open(file) as f:
        lines = f.readlines()
    
        for line in lines:
            
            if line == '\n':
                break
                
            else:
                line=line.strip()
                #print(line,len(line),current,'last:',line[-1])
                if line=='$ cd /':
                    #print('root node')
                    current=root
                    
                    listing=False
                    continue
                
                # check of $ls
                elif line=='$ ls':
                    #print('listing')
                    listing=True
                #  # go to parent node   
                elif line=='$ cd ..':
                    
                    #print('go to parent')
                    #print('current',current)
                    current=current.parent

                    listing=False

                # move to subdirectory  
                elif (line[:4]=='$ cd' and line[-2:]!='..' and len(line)>=6):
                    #print('--node change')
                    new_node_name=line.split(' ')[2]
                    # checks if parent exists
                    curr=current
                    if current.parent is not None:
                        current=find(root,lambda node:node.parent==curr and node.name==new_node_name)
                    else:
                        current=find(root,lambda node:node.parent==root and node.name==new_node_name)
                    #print('new curr:',current)
            
                    listing=False
                else:
                    #print('list children')
                    if listing is True:
                        file_spec=line.split(' ')
                            
                        
                        if line[:4]!='dir ':
                            file_name,size=file_spec[1],int(file_spec[0])
                            type='file'
                        else:
                            size=0
                            file_name=file_spec[1]
                            type='dir'

                        #print(current,"file_name:",file_name,'size:',size)
                            
                        # node_loc=find(root,lambda node:node.name==file_name)
                        # if node_loc==None:
                        #     print('---new node')
                        #     temp=FileNode(file_name,size=size,type=type,parent=current)
                        #     print(temp)
                        # elif node_loc!=None and next(node_loc.ancestors) is current:
                        #     print('---new node')
                        #     temp=FileNode(file_name,size=size,type=type,parent=current)
                        #     print(temp)
                        temp=FileNode(file_name,size=size,type=type,parent=current)

                    else: continue

    

    #print_tree(root,attr_list=['size'])
    return root
```

```python
def delete_directory(tree):
    nodes=findall(tree,lambda node:node.type=='dir')
    total=0
    size_dict={}
    for node in nodes:
        if node.name=='root':
            size_dict['root']=node.size
            size_dict['unused']=70000000-node.size
            size_dict['size_to_delete']=30000000-size_dict['unused']
        else: 
            total+=node.size
    size_dict['used']=total

    arr=[node.size for node in nodes if node.size>=size_dict['size_to_delete']]

    return min(arr)

delete_directory(tree)
```
