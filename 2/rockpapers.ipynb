```python
rock=['A','X']
paper=['B','Y']
scissors=['C','Z']

scores={'rock':1,
        'paper':2,
        'scissors':3}

        

```

```python
def score(opponent,me):

    if( me in rock and opponent in rock) or( me in paper and opponent in paper) or ( me in scissors and opponent in scissors):

        

        if(me in rock):
            return scores['rock']+3
        elif(me in paper):
            return scores['paper']+3
        elif(me in scissors):
            return scores['scissors']+3
    # rock
    elif me in rock and opponent in scissors:
        return scores['rock']+6
    elif me in rock and opponent in paper:
        return scores['rock']+0

    # paper
    elif me in paper and opponent in rock:
        return scores['paper']+6
    elif me in paper and opponent in scissors:
        return scores['paper']+0
    
    # scissors
    elif me in scissors and opponent in paper:
        return scores['scissors']+6
    elif me in scissors and opponent in rock:
        return scores['scissors']+0
    





```

```python
def get_scores(file='sample.txt'):
    round_count=0
    
    total_score=0
    
    with open(file) as f:
        lines = f.readlines()

        

        
        for line in lines:
            
            if line == '\n':
                round_count+=1
                
            else:

                choices=line.strip().split(' ')
                
               
                total_score=total_score+score(choices[0],
                                                    choices[1]
                                                )
                print(choices,total_score)

            
    
    return total_score
```
