```python
def check_complete_overlap(pairA,pairB):

    A_vals = pairA.split('-')
    B_vals = pairB.split('-')

    # print(pairA,pairB)
    # print(A_vals,B_vals)
    startA,endA=int(A_vals[0]),int(A_vals[1])
    startB,endB=int(B_vals[0]),int(B_vals[1])

    

    similarity_counter=0
    if(startA==endA):
        similarity_counter+=1
    if(startB==endB):
        similarity_counter+=2

    print(startA,endA,startB,endB,similarity_counter)
    
    if(similarity_counter==3 and startA==startB):
        return True

    
    # A_vals same so iterating over B    
    elif(similarity_counter==1):
        if (startA>=startB and startA<=endB):
            return True
        else:
            return False

    # B_vals same so iterating over A    
    elif(similarity_counter==2):
        if startB>=startA and startB<=endA:
            return True
        else:
            return False

    # no similarity
    else:

        #check for overlapping

        #check if startA and endA between startB and endB
        if startA>=startB and endA<=endB:
            return True
        #check if startA and endA between startB and endB
        elif startB>=startA and endB<=endA:
            return True
        else:
            return False



    


    
```

```python
def check_partial_overlap(pairA,pairB):

    A_vals = pairA.split('-')
    B_vals = pairB.split('-')

    # print(pairA,pairB)
    # print(A_vals,B_vals)
    startA,endA=int(A_vals[0]),int(A_vals[1])
    startB,endB=int(B_vals[0]),int(B_vals[1])

    

    similarity_counter=0
    if(startA==endA):
        similarity_counter+=1
    if(startB==endB):
        similarity_counter+=2

    print(startA,endA,startB,endB,similarity_counter)
    
    if(similarity_counter==3 and startA==startB):
        return True

    
    # A_vals same so iterating over B    
    elif(similarity_counter==1):
        if (startA>=startB and startA<=endB):
            return True
        else:
            return False

    # B_vals same so iterating over A    
    elif(similarity_counter==2):
        if startB>=startA and startB<=endA:
            return True
        else:
            return False

    # no similarity
    else:

        #check for overlapping

        #check if startA and endA between startB and endB
        if startA>=startB and endA<=endB:
            return True
        #check if startA and endA between startB and endB
        elif startB>=startA and endB<=endA:
            return True

        # 5-7,7-9
        elif (startA>=startB and endA<=endB) or (startB>=startA and endB<=endA):
            return True
        
        elif (endA>=startB and endA<=endB) or (endB>=startA and endB<=endA):
            return True
        else:
            return False



    


    
```
