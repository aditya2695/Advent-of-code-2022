```python
def tuple_sum(tup1,tup2,op='sum'):
    #print('tuples:',tup1,tup2)
    if op=='sum':
        res = tuple(map(lambda i, j: i + j, tup1, tup2))
    else:
        res = tuple(map(lambda i, j: i - j, tup1, tup2))
    return res        
print((tuple_sum((5,3),(2,2),'sum')))
print(tuple_sum((5,3),(2,2),'diff'))
```

```python
# def checkAdjacency(H,T):
#     if H==T:
#         return True
#     Hx,Hy,Tx,Ty = H[0],H[1],T[0],T[1]
    
    
#     if Hx==Tx or Hy==Ty:
#         return True
#     # postive top diagonal
#     elif Hx-Tx==1 and Hy-Ty==1:
#         return True
#     # negative top diagonal
#     elif Hx-Tx==-1 and Hy-Ty==1:
#         return True
#     elif Hx-Tx==-1 and Hy-Ty==-1:
#         return True
#     elif Hx-Tx==1 and Hy-Ty==-1:
#         return True
#     else:
#         return False

def tail_change(T,H):

    print(H,T)
    if H==T:
        #print('overlap')
        return (0,0)
        

    Hx,Hy,Tx,Ty = H[0],H[1],T[0],T[1]

    # same column but adjacent 
    if Hx==Tx and abs(Hy-Ty)==1:
        return (0,0)
    # same column but not adjacent 
    elif Hx==Tx and abs(Hy-Ty)>1:
        return (0,1) if Hy-Ty>0 else (0,-1)
    # same row but adjacent 
    elif Hy==Ty and abs(Hx-Tx)==1:
        return (0,0)
    # same row but not adjacent 
    elif Hy==Ty and abs(Hx-Tx)>1:
        return (1,0) if Hx-Tx>0 else (-1,0)

    # move to postive right diagonal - col change
    elif tuple_sum(T,(1,2),op='sum')==H:
        return (1,1)
    # move to postive right diagonal -- row change
    elif tuple_sum(T,(2,1),op='sum')==H:
        return (1,1)

    # move to postive left diagonal - col change
    elif tuple_sum(T,(-1,2),op='sum')==H:
        return (-1,1)
    # move to postive left diagonal -- row change
    elif tuple_sum(T,(-2,1),op='sum')==H:
        return (-1,1)
    
    # move to negative right diagonal - col change
    elif tuple_sum(T,(1,-2),op='sum')==H:
        return (1,-1)
    # move to negative right diagonal-- row change
    elif tuple_sum(T,(2,-1),op='sum')==H:
        return (1,-1)

    # move to negative left diagonal - col change
    elif tuple_sum(T,(-2,-1),op='sum')==H:
        return (-1,-1)
    # move to negative left diagonal-- row change
    elif tuple_sum(T,(-1,-2),op='sum')==H:
        return (-1,-1)
    
    
    else:
        return (0,0)




    
```

```python
class Rope:
    def __init__(self) -> None:
        self.H=(0,0)    # (x,y)
        self.T=(0,0)    # (x,y)
        self.tail_pos=set()

    def move(self,direction,steps):

        for i in range(1,steps+1):
            if direction=='R':
                self.H=tuple_sum(self.H,(1,0),op='sum')

                

            elif direction=='L':
                #self.H=self.H-(steps,0)
                self.H=tuple_sum(self.H,(1,0),op='diff')
            elif direction=='D':
                #self.H=self.H-(0,steps)
                self.H=tuple_sum(self.H,(0,1),op='diff')
            elif direction=='U':
                #self.H=self.H+(0,steps)
                self.H=tuple_sum(self.H,(0,1),op='sum')

            tail_step  =tail_change(self.T,self.H)
            self.T=tuple_sum(self.T,tail_step,op='sum')
            #print(direction,i,self.H,self.T,'\n')

            self.tail_pos.add(self.T)

        return self.H,self.T

    

```
